```python
# Cargar librería completa con un nombre

import pydicom as pyd

file_dataset=pyd.dcmread("intrinsic_unif")

file_dataset
```

```python
# Cargar solo un módulo de la librería

from pydicom import dcmread

file_dataset=dcmread("intrinsic_unif")

file_dataset
```

```python
# Tipos de objeto

type(file_dataset)
```

```python
# Acceder datalement mediante tag

selected_tag=file_dataset[0x00100030]

print(selected_tag)

type(selected_tag)
```

```python
print("Tag: "+ str(selected_tag.tag))

print("Name: "+ str(selected_tag.name))

print("Value representation: "+selected_tag.VR)

print("Value multiplicity: "+ str(selected_tag.VM))

print("Value representation: "+ str(selected_tag.value))



```

```python
import os

i=0

for root,dirs,files in os.walk("ct-phantom"):
    for name in files:
        
        if i<10:
            name_start=str(0)+str(i)
            
        else:
            name_start=str(i)
            
        name_end="-file"
        
        name_new=name_start+name_end
        
        os.rename(os.path.join(root,name), os.path.join(root,name_new))
        
        i+=1
        
for root,dirs,files in os.walk("ct-phantom"):
    for name in files:
        print(name)
```
